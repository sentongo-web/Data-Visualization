## Exercise 2: Measures of Central Tendency

### Mean

```python
# Function to calculate Mean
def calculate_mean(x):
    if len(x) == 0:
        raise ValueError("List is empty, cannot calculate mean!")
    total = sum(x)
    mean = total / len(x)
    return mean
```

```python
calculate_mean(rates_numeric)
```

##### Most of the rates are grouped together around 1383.34.

### Median

```python
# Function to calculate Median
def calculate_median(x):
    if len(x) == 0:
        raise ValueError("List is empty, cannot calculate median!")
    sorted_x = sorted(x)
    n = len(sorted_x)

    if n % 2 == 0:
        # For an even-sized list, average the middle two elements
        middle_left = sorted_x[n // 2 - 1]
        middle_right = sorted_x[n // 2]
        median = (middle_left + middle_right) / 2
    else:
        # For an odd-sized list, return the middle element
        median = sorted_x[n // 2]
    return median
```

```python
calculate_median(rates_numeric)
```

##### With a median value of 1196.34, half of the data points are below and half are above this figure. This acts as a central reference point, displaying the values' center position and illuminating the central tendency of the distribution.

### Mode

```python
from collections import Counter


# Function to calculate Mode
def calculate_mode(x):
    if len(x) == 0:
        raise ValueError("List is empty, cannot calculate mode!")
    # Use Counter to count occurrences of each element
    counts = Counter(x)

    # Find the maximum count
    max_count = max(counts.values())

    # Get all elements with maximum count (multiple modes are possible)
    modes = [key for key, value in counts.items() if value == max_count]

    if len(modes) == 1:
        # If there's only one mode, return it
        return modes[0]
    else:
        # If there are multiple modes, return a list of modes
        return modes
```

### Interquartile Range (IQR)

```python
def calculate_iqr(i):
    if len(i) < 4:
        raise ValueError("IQR requires at least four data points.")
    sorted_data = sorted(i)

    # Calculate Q1 and Q3
    q1_index = (len(sorted_data) - 1) // 4
    q3_index = 3 * (len(sorted_data) - 1) // 4

    q1 = (sorted_data[q1_index] + sorted_data[q1_index + 1]) / 2
    q3 = (sorted_data[q3_index] + sorted_data[q3_index + 1]) / 2

    # Calculate IQR
    iqr = q3 - q1
    return iqr
```

```python
calculate_iqr(rates_numeric)
```

#### Coefficient of Variation:

```python
def calculate_coefficient_of_variation(i):
    if len(i) == 0:
        raise ValueError("List is empty, cannot calculate CV!")
    mean_value = sum(i) / len(i)
    std_dev = (sum((x - mean_value) ** 2 for x in i) / len(i)) ** 0.5
    coefficient_of_variation = (std_dev / mean_value) * 100
    return coefficient_of_variation
```

```python
calculate_coefficient_of_variation(rates_numeric)
```

##### A significant degree of variability in relation to the mean is indicated by a coefficient of variation value of 90.41

#### Quartile Coefficient of Dispersion

```python
def calculate_quartile_coefficient_of_dispersion(i):
    if len(i) < 4:
        raise ValueError("QCD requires at least four data points.")
    sorted_data = sorted(i)
    q1_index = (len(sorted_data) - 1) // 4
    q3_index = 3 * (len(sorted_data) - 1) // 4
    q1 = (sorted_data[q1_index] + sorted_data[q1_index + 1]) / 2
    q3 = (sorted_data[q3_index] + sorted_data[q3_index + 1]) / 2
    qcd = (q3 - q1) / (q3 + q1)
    return qcd
```

```python
calculate_quartile_coefficient_of_dispersion(rates_numeric)
```

##### The gap between the first and third quartiles is less than the entire range of the data (IQR < Range), as indicated by the QCD value of 0.914

```python
# Since statistics are applied to numeric values, I will convert 'rates' to numeric values
coded_dates_numeric = pd.to_numeric(coded_dates, errors='coerce')
```

```python
def calculate_covariance(k, l):
    if len(k) != len(l) or len(k) < 2:
        raise ValueError(
            "Input data should have the same length and at least two data points.")
    mean_k = sum(k) / len(k)
    mean_l = sum(l) / len(l)
    covariance = sum((k[i] - mean_k) * (l[i] - mean_l) for i in
                     range(len(k))) / (len(k) - 1)
    return covariance
```

```python
calculate_covariance(rates_numeric, coded_dates_numeric)
```

##### With a covariance value of about -29452.27, the two variables are negatively correlated.

#### Pearson Correlation Coefficient:

```python
def calculate_pearson_correlation(k, l):
    if len(k) != len(l) or len(k) < 2:
        raise ValueError(
            "Input data should have the same length and at least two data points.")

    covariance = calculate_covariance(k, l)
    std_dev_k = (sum((x - sum(k) / len(k)) ** 2 for x in k) / len
    (k)) ** 0.5
    std_dev_l = (sum((x - sum(l) / len(l)) ** 2 for x in l) / len
    (l)) ** 0.5
    pearson_correlation = covariance / (std_dev_k * std_dev_l)
    return pearson_correlation
```

```python
calculate_pearson_correlation(rates_numeric, coded_dates_numeric)
```

##### There appears to be a slight negative linear relationship between dates  and rates, as indicated by the Pearson correlation coefficient of roughly -0.13.

#### Percentiles:

```python
def calculate_percentile(i, percentile):
    if len(i) < 1 or percentile < 0 or percentile > 100:
        raise ValueError("Invalid input data or percentile value.")

    sorted_i = sorted(i)
    index = (percentile / 100) * (len(sorted_i) - 1)

    if index.is_integer():
        return sorted_i[int(index)]
    else:
        lower_index = int(index // 1)
        upper_index = lower_index + 1
        lower_value = sorted_i[lower_index]
        upper_value = sorted_i[upper_index]
        return lower_value + (index - lower_index) * (upper_value - lower_value)
```

```python
tenth_percentile = calculate_percentile(rates_numeric, 10)
second_percentile = calculate_percentile(rates_numeric, 25)
third_percentile = calculate_percentile(rates_numeric, 50)
fourth_percentile = calculate_percentile(rates_numeric, 75)

print(f' 10th Percentile: {tenth_percentile}\n 25th Percentile: '
      f'{second_percentile}\n 50th Percentile: '
      f'{third_percentile}\n 75th Percentile: {fourth_percentile}\n')
```

#### Skewness

```python
def calculate_skewness(i):
    if len(i) < 3:
        raise ValueError("Skewness requires at least three data points.")

    mean_value = sum(i) / len(i)
    std_dev = (sum((x - mean_value) ** 2 for x in i) / len(i)) ** 0.5
    skewness = sum(((x - mean_value) / std_dev) ** 3 for x in i) * \
               (len(i) / ((len(i) - 1) * (len(i) - 2)))
    return skewness
```

```python
calculate_skewness(rates_numeric)
```

##### The data distribution has a somewhat positive skewness, as indicated by the skewness value of 0.541
